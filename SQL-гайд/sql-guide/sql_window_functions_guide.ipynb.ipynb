```python
import pandas as pd

# Турнирная таблица
tournament = pd.DataFrame({
    'player': ['Игрок1', 'Игрок2', 'Игрок3', 'Игрок4', 'Игрок5', 'Игрок6'],
    'score': [100, 90, 90, 80, 80, 70]
})

# Добавляем ранги
tournament['row_number'] = tournament['score'].rank(method='first', ascending=False).astype(int)
tournament['rank'] = tournament['score'].rank(method='min', ascending=False).astype(int)
tournament['dense_rank'] = tournament['score'].rank(method='dense', ascending=False).astype(int)

print("ТУРНИРНАЯ ТАБЛИЦА")
print(tournament)
```

## Разница между функциями

- **ROW_NUMBER** - уникальная нумерация (1,2,3,4,5,6)
- **RANK** - места с пропусками (1,2,2,4,4,6)
- **DENSE_RANK** - места без пропусков (1,2,2,3,3,4)

## Итог

Все три функции - мощный инструмент аналитика.
Главное - понимать бизнес-задачу и выбирать правильную.
